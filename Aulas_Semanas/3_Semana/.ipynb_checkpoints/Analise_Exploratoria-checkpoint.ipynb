```python
aux = pd.DataFrame({'colunas' : dados.columns, 'tipos' : dados.dtypes})
```

```python
colunas_numericas = list(aux[aux['tipos'] == 'int64']['colunas'])
```

```python
for coluna in colunas_numericas:
    #FaÃ§a correlaÃ§Ãµes entre valor do aluguel e cada coluna
    #Logo precisei pegar quais sÃ£o as colunas que contÃ©m valores numÃ©ricos
    #Desta forma conseguis printar as correlaÃ§Ãµes na forma de tabela
    print(dados[['valor_aluguel', coluna]].corr(method = 'spearman'))
```

```python
plt.figure(figsize=(12,6))
sea.barplot(x = 'city', y = 'valor_aluguel', data = dados.groupby('city')['valor_aluguel'].mean().reset_index())
```

```python
#the geom_col must be used if I want to attribute a value to y axis, instead of the geom_bar which uses count stats
(ggplot(dados.groupby('city')['valor_aluguel'].mean().reset_index()) + 
 aes(x = 'city', y='valor_aluguel') + 
 geom_col(aes(fill = 'city')) + 
 ggtitle('MÃ©dia dos valores de aluguÃ©is por cidade') + 
 theme_bw())
```
