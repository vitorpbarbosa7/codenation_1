### Shapiro Wilk

```python
normal_data = sct.norm.rvs(loc = 10, scale = 4, size =100)
```

```python
sct.shapiro(normal_data) 

#Segundo valor é o valor do p, ou seja, a probabilidade de que ao 
#rejeitar Ho eu esteja errado 

#Se a hipótese nula é de Ho: os dados são normais~
#e H1: dados não são normais

#temos que probabilidades grandes indicam que ao rejeitar Ho, estou rejeitando a normalidade, e estou com alta probabilidade estar rejeitando erroneamente

#Com probabilidades pequenas, estou com probabilidade de rejeitar Ho e não errar
```

```python
sm.qqplot(normal_data, fit  = True, line = '45')
```
