```python
from math import sqrt

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import scipy.stats as sct
import seaborn as sns
from sklearn.datasets import load_digits
```

```python
from sklearn.decomposition import PCA
```

```python
#Dataset digits
```

```python
digits = load_digits()
```

```python
df_digits = pd.DataFrame(digits.data)
```

```python
#Inicialização do objeto PCA
pca = PCA(n_components=2)
```

```python
projected = pca.fit_transform(digits.data)
```

```python
print(f"Original Shape: {digits.data.shape},projected shape: {projected.shape}")
```

```python
sns.scatterplot(projected[:,0],projected[:,1])
```

```python
pca = PCA().fit(digits.data)
```

```python
evr = pca.explained_variance_ratio_
```

```python
g = sns.lineplot(np.arange(len(evr)),
                np.cumsum(evr))
g.axes.axhline(0.95, ls = "--", color="red")
plt.xlabel('Número de componentes')
plt.ylabel('Variância explicada acumulada')
```

#### Número exato de componentes para explicar 95 % da variância

#### Diferencial da variância explicada

```python
sns.lineplot(np.arange(len(evr)), evr)
```

```python
variancia_explicada_acumulada = np.cumsum(pca.explained_variance_ratio_)
```

```python
sns.lineplot(np.arange(len(variancia_explicada_acumulada)), variancia_explicada_acumulada)
```

```python
#Retorna o número do índice que representa mais que 0.95 de variância acumulada
np.argmax(variancia_explicada_acumulada >= 0.95)
```

```python
#Retorna o índice do maior elemento de um vetor em um determinado eixo´
numero_componentes = np.argmax(variancia_explicada_acumulada >= 0.95) + 1; numero_componentes
```
