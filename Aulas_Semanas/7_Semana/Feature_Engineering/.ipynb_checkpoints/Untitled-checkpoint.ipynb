## Discretização (Binning)

#### Utilizando quantile:

Desta maneira, defino quartis que possuem aproximadamente a mesma quantidade de observações

```python
#Discretizer é o meu objeto que recebeu o fit
discretizer = KBinsDiscretizer(n_bins = 4, encode = 'ordinal', 
                              strategy = 'quantile')

discretizer.fit(data_encoded[['Score']])

#score_bins é o resultado do transform aplicado pelo discretizer sobre uma das colunas, nesse caso a coluna 'score'
score_bins = discretizer.transform(data_encoded[['Score']])

score_bins[:10]
```

```python
plt.hist(score_bins)
```

Limites dos intervalos disponíveis em bin_edges_

```python
discretizer.bin_edges_[0]
```

Criando as labels que indicam os intervalos:

```python
def get_interval(bin_idx, bin_edges):
    return f"{np.round(bin_edges[bin_idx], 2):.2f} ⊢ {np.round(bin_edges[bin_idx+1],2):.2f}"
```

```python
#Testando para ver como funciona essa função
get_interval(1, discretizer.bin_edges_[0])
```

```python
#Armazenar em uma lista os intervalos
lista_intervalo = []

for i in range(len(discretizer.bin_edges_[0])-1):
    lista_intervalo.append(get_interval(i, discretizer.bin_edges_[0]))
```

```python
lista_intervalo
```

```python
score_intervals = pd.Series(score_bins.flatten().astype(np.int)).apply(get_interval,args = (discretizer.bin_edges_[0],))
```

# Outliers

```python
#Função copy que permite (obviamente) a cópia dos dados
height_outlier = data.Height.copy()
type(height_outlier)
```

Adicionamos outliers:

```python
height_outlier_idx = pd.Index(np.random.choice(height_outlier.index, 10, replace = False))

#Os primeiros cinco índices serão de too short:
too_short_idx = pd.Index(height_outlier_idx[:5])
too_tall_idx = pd.Index(height_outlier_idx[5:])

#Preenchendo os índices com valores gerados a partir de uma normal:
height_outlier[too_short_idx] = np.random.normal(loc = 1.3, scale = 0.5, size = 5)
height_outlier[too_tall_idx] = np.random.normal(loc = 2.20, scale = 0.5, size = 5)
```

```python
#Para pegar os índices dos too_short e too_tall:
#Pipe character '|' é o operador de união
outliers_idx = too_short_idx | too_tall_idx
type(outliers_idx)
```

```python
height_outlier[outliers_idx]
```

#### Boxplot

```python
sns.boxplot(height_outlier, orient = 'vertical')
```

#### Devemos agora retornar uma lista que contenha todos os outliers de acordo com o critério de quartis do boxplot

```python
#Quartis superior e inferior
q1 = height_outlier.quantile(0.25)
q3 = height_outlier.quantile(0.75)
#Região interquartis
iqr = q3 - q1

outlier_limits = [q1 - 1.5*iqr, q3 + 1.5*iqr]

outliers = height_outlier[(height_outlier < outlier_limits[0]) | (height_outlier > outlier_limits[1])]

#É interessante observar que como este objeto outliers foi obtido a partir da série do pandas 'height_outliers' ele vem acompanhado dos respectivos
#índices

outliers
```

```python
#Para retirar os outliers:
height_outlier_drop = height_outlier.drop(index = outliers.index)
```

```python
#Pode-se visualizar no boxplot como agora não há mais outliers
sns.boxplot(height_outlier_drop, orient = 'vertical')
```

### Testes de hipótese para detectar outliers

<font color = 'red' size = 6> Teste de Grubb </font>

```python
#Criação das funções:
def grubb_test(g, n, alpha=0.05, tailed='two-tailed'):
  if tailed == 'two-tailed':
    critical = ((n - 1)/sqrt(n)) * sqrt(sct.t.isf(alpha/(2*n), n-2)**2/(n - 2 + sct.t.isf(alpha/(2*n), n-2)**2))
    
    return (g, critical, g > critical)
  elif tailed == 'one-tailed':
    critical = ((n - 1)/sqrt(n)) * sqrt(sct.t.isf(alpha/(n), n-2)**2/(n - 2 + sct.t.isf(alpha/(n), n-2)**2))
    
    return (g, critical, g > critical)
  else:
    raise ValueError(f"Invalid tailed argument")
                                         
```

```python
def grubb_summary(result, decimals=10):
  return (
    f"Null hypothesis: there is no outliers in the data set\n"
    f"Test statistic: {np.round(result[0], decimals)}, "
    f"Grubb's critical value: {np.round(result[1], decimals)}, "
    f"Reject: {result[2]}"
  )
```

```python
def next_outlier_candidate(data):
  sample_distances = (data - data.mean()).abs()
  candidate_idx = sample_distances.idxmax()
  candidate_value = data[candidate_idx]
  candidate_statistic = sample_distances.max()/data.std()
  
  return (candidate_idx, candidate_value, candidate_statistic, len(data))
```

## São exibidos os valores no quais a hipótese nula foi rejeitada, ou seja, há outliers

```python
height_outlier_grubb = height_outlier.copy()

outliers_grubb = pd.Series()

has_outlier = True

while has_outlier == True:
    outlier_candidate = next_outlier_candidate(height_outlier_grubb)
    
    print(f"Index: {outlier_candidate[0]}), "
          f"Value: {np.round(outlier_candidate[1], 3)}, "
          f"Test Statistic: {np.round(outlier_candidate[2], 3)}, "
          f"Sample size: {outlier_candidate[3]}\n")

    result = grubb_test(outlier_candidate[2], outlier_candidate[3])
    
    print(grubb_summary(result, 3))

    has_outlier =  result[2]

    if has_outlier:
        height_outlier_grubb = height_outlier_grubb.drop(index = outlier_candidate[0])
        outliers_grubb.at[outlier_candidate[0]] = outlier_candidate[1]

    print(f"\n")
```
